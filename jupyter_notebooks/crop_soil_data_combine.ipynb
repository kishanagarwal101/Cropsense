# Combining Crop and Soil Data into one csv

```python
soil_merged_data_path = '../data/data_raw/soil_data_merged.csv'
soil_merge = pd.read_csv(soil_merged_data_path)
```

```python
crop_merge = pd.read_csv('../data/data_raw/crop_data_merged.csv')
reco_soil = soil_merge
```

```python
#Add +/-3 for every NPK value
import random
import pandas as pd

temp_data = [
    {
        "N": reco_soil.loc[reco_soil['Crop'] == crop, 'N'].iloc[0] + random.randint(-20, 20),
        "P": reco_soil.loc[reco_soil['Crop'] == crop, 'P'].iloc[0] + random.randint(-5, 20),
        "K": reco_soil.loc[reco_soil['Crop'] == crop, 'K'].iloc[0] + random.randint(-5, 5)
    }
    for crop in crop_merge.label
]

temp = pd.DataFrame(temp_data, columns=['N', 'P', 'K'])
```

```python
temp
```

```python
crop_merge['N'] = temp['N']
crop_merge['P'] = temp['P']
crop_merge['K'] = temp['K']
```

```python
crop_merge
```

```python
del crop_merge['Unnamed: 0']
```

```python
crop_merge = crop_merge[[ 'N', 'P', 'K','temperature', 'humidity', 'ph', 'rainfall', 'label']]
```

```python
crop_merge
```

```python
crop_merge.to_csv("../data/data_processed/crop_recommendation_merged_dataset.csv" , index=False)
```
